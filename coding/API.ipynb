### 1. REST API: Number of Drawn Matches
In this challenge, the REST API contains information about football matches. The provided API allows querying matches by teams and year. The task is to get the number of matches for a given year that ended in a draw. A match is drawn when both teams scored the same number of goals.

To access a collection of matches played in a given year, perform an HTTP GET request to https://jsonmock.hackerrank.com/api/football_matches?year=syear>&page=<page>
where ‹year> is the year of the competition and <page> is the page of the results to request. The results might be divided into several pages. Pages are numbered from 1.

For example, a GET request to
https://jsonmock.hackerrank.com/api/football_matches?year=2011&page=2 returns data associated with matches in the year 2011 on the second page of the results.

The response to such a request is a SON with the following 5 fields:
• page: The current page of the results.
• per_page: The maximum number of matches returned per page.
• total: The total number of matches on all pages of the results.
• total_pages: The total number of pages with results.
• data: An array of objects containing matches information on the requested page.
Each match record has several fields:
• competition: a string denoting the name of the competition
• year: an integer denoting the year when the match took place
• round: a string denoting the round the match belongs to (can be an empty string) team: a string denoting the name of the first team in the match
• team2: a string denoting the name of the second team in the match
• team1goals: a string denoting the number of goals scored by team1 in the match
• team2goals: a string denoting the number of goals scored by team2 in the match

Notice that the number of pages might be in hundreds, and it would take too much time to fetch the results from all of them and examine the scores of every match. In order to overcome this issue, you are allowed to add an exact value of any of the match object fields to the URL query string in order to limit the number of results. This capability, if used correctly, can help you avoid examining individual match objects.

For example, performing a request to
https://jsonmock.hackerrank.com/api/football_matches?year=2011&team1goals=1&page=2 returns data associated with matches in the year 2011, where the first team scored 1 goal, on the second page of the results.

```python
import requests

def getNumDraws(year):
    url = "https://jsonmock.hackerrank.com/api/football_matches"
    total_draws = 0

    for goal in range(11):
        params = {
            "year": year,
            "team1goals": goal,
            "team2goals": goal,
            "page":1
        }
        response = requests.get(url, params=params)
        data = response.json()

        total_draws = total_draws + data["total"]
    return total_draws

print(getNumDraws(2011))
```

```python
def getCompetitionWinner(name, year):
    url = "https://jsonmock.hackerrank.com/api/football_competitions"
    params = {
        "name": name,
        "year": year
    }
    response = requests.get(url, params=params)
    data = response.json()["data"][0]
    print("data:", data)
    return data["winner"]
# name, country , year, winner, runner up
# team1 - home team
# team2 - visiting team
def getTeamGoals(competition, year, team, side):
    url1 = "https://jsonmock.hackerrank.com/api/football_matches"
    total_goals = 0
    page = 1
    while True:
        params = {
            "competition":competition,
            "year":year,
            side: team,
            "page": page
        }
        response = requests.get(url1, params=params)
        data = response.json()
        total_pages = data["total_pages"]
        for match in data["data"]:
            if side == "team1":
                total_goals = total_goals + int(match["team1goals"])
            else:
                total_goals = total_goals + int(match["team2goals"])
        if page >= data["total_pages"]:
            break
        page = page + 1
    return total_goals
# for a given competition and year return total number of goals scored by the team who won the competition

def getWinnerGoals(competition, year):
    winner = getCompetitionWinner(competition, year)
    goals1 = getTeamGoals(competition, year, winner, "team1")
    goals2 = getTeamGoals(competition, year, winner, "team2")
    return goals1 + goals2
```
