### Problem Statement
You were required to implement three sub-routines using the producer–filter–consumer pattern to process a stream of numbers.
The order of operations was dynamic and provided at runtime.
Sub-routines to Implement
1. accumulator() - Maintains a running sum of the received numbers
2. squarer() - Outputs the square of the received number
3. rooter() - Outputs the floor value of the square root of the received number

```python
import math
from collections.abc import Generator

def accumulator(source: Generator) -> Generator:
    res_sum = 0
    for value in source:
        res_sum = res_sum + value
        yield res_sum

def squarer(source: Generator) -> Generator:
    for value in source:
        yield value**2

def rooter(source: Generator) -> Generator:
    for value in source:
        yield math.floor(math.sqrt(value))
    pass

def produce_numbers(nums:list) -> Generator:
    for num in nums:
        yield num
        
def build_pipeline(nums:list, operations:list[str])-> Generator:
    op_map = {
        "accumulator":accumulator,
        "squarer":squarer,
        "rooter":rooter
    }
    pipeline = produce_numbers(nums)
    for op_name in operations:
        pipeline = op_map[op_name](pipeline)
        
    return pipeline
    
    
if __name__ == "__main__":
    nums = [1,2,3,4,5]
    operations = ["squarer","accumulator"]
    res = list(build_pipeline(nums, operations))
    print(res)
```

### Problem : Lambda Sort (Multi-level Sorting)
Problem Statement:

You were required to complete a lambda expression inside the indexSort function to sort a 2D array based on multiple columns and sorting directions.
Key Requirements
1. Sorting must be stable
2. Multiple keys and directions must be handled correctly
    Direction 1 indicates descending order
    Direction 0 indicates ascending order
Example
If indices = [[1, 0], [2, 1]]
Then the array should be sorted:
    Primarily by column 1 (ascending)
    Secondarily by column 2 (descending) in case of ties

```python
from typing import List

def index_sort(data: List[List[int]], indices: List[List[int]]) -> List[List[int]]:
    sorted_data = list(data)
    
    for col_idx, direction in reversed(indices):
        is_descending = (direction == 1)
        sorted_data.sort(key=lambda row:row[col_idx], reverse=is_descending)
        
        
    return sorted_data

if __name__ == "__main__":
    # Sample Dataset matching your rules: [ID, Column 1, Column 2]
    # Rows 1, 3, and 5 tie on Column 1 (value 25).
    # Rows 2 and 4 tie on Column 1 (value 30).
    sample_data = [
        [1, 25, 85],  # Tie-breaker should put this first among 85s (Stability)
        [2, 30, 95],
        [3, 25, 90],
        [4, 30, 85],
        [5, 25, 85]   # Tie-breaker should keep this after row ID 1
    ]
    
    # Rules: 
    # Primary: Column 1 (Ascending)
    # Secondary: Column 2 (Descending)
    sorting_rules = [[1, 0], [2, 1]]
    
    result = index_sort(sample_data, sorting_rules)
    
    print("Final Sorted Matrix:")
    for row in result:
        print(row)
```

### Problem statement: number of odd sized subarrays with given median
given an array and an index , return the number of odd length sub-arrays that have the median value equal to the value at given input index , array.=[2,3,5,4,5,3,2] index = 1

```python
def count_odd_median_subarrays_fast(arr, index):
    target = arr[index]
    n = len(arr)
    count = 0
    for i in range(n):
        less = equal = greater = 0
        for j in range(i, n):
            v = arr[j]
            if v < target:
                less += 1
            elif v > target:
                greater += 1
            else:
                equal += 1
            length = j - i + 1
            if length % 2 == 1 and abs(greater - less) <= equal - 1:
                count += 1
    return count
```
