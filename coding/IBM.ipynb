### Problem Statement: The Value-Based Node Expansion Tree
Consider an infinite, rooted tree where the vertices are labeled sequentially with positive integers starting from 1 at the root.

1. Level 1: Contains the root node 1. It has 1 child.
2. Level 2: Contains node 2 (the child of 1). It has 2 children.
3. Level 3: Contains nodes 3 and 4 (the children of 2). Node 3 spawns 3 children, and Node 4 spawns 4 children.
4. Level 4: Contains nodes 5, 6, 7 (children of 3) and nodes 8, 9, 10, 11 (children of 4).

Given an input integer K (K>1), determine the label of its parent node.
