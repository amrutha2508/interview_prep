```python
# Turing python question
# given a list of numbers (0's and 1's) return the length of the longest sequence

def longest_alternating_subsequence(ml):
    if not ml:
        return 0, []

    max_len = 1
    start_idx = 0
    best_start = 0

    current_len = 1

    for i in range(1, len(ml)):
        # Check if the current element alternates with the previous one
        if ml[i] != ml[i - 1]:
            current_len += 1
        else:
            # If they don't alternate, check if this was the longest run so far
            if current_len > max_len:
                max_len = current_len
                best_start = start_idx

            # Reset the window to start at the current element
            current_len = 1
            start_idx = i

    # Final check in case the longest sequence goes all the way to the end
    if current_len > max_len:
        max_len = current_len
        best_start = start_idx

    return max_len, ml[best_start : best_start + max_len]


# --- Test Cases ---
print(
    "Test [1, 0, 0, 0]:", longest_alternating_subsequence([1, 0, 0, 0])
)  # Longest alternating is [1, 0] (length 2)
print(
    "Test [0, 1, 0, 1, 0]:", longest_alternating_subsequence([0, 1, 0, 1, 0])
)  # Odd length: (5, [0, 1, 0, 1, 0])
print(
    "Test [1, 1, 0, 1, 0, 0]:",
    longest_alternating_subsequence([1, 1, 0, 1, 0, 0]),
)  # Even length mixed: (4, [1, 0, 1, 0])
print(
    "Test [0, 0, 0]:", longest_alternating_subsequence([0, 0, 0])
)  # No alternation: (1, [0])
```
